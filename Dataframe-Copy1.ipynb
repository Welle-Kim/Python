```python
import re,string
```

```python
from konlpy.tag import Okt
okt=Okt()
punc_regex = re.compile('[{}]'.format(re.escape(string.punctuation)))
```

```python
noun_list=[]
verb_list=[]
josa_list=[]

for w in df_인식결과["Script"]:
    for word in okt.pos(w):
        if word[1] in ["Noun"]:
            noun_list.append(word[0])
        if word[1] in ["Verb"]:
            verb_list.append(word[0])
        if word[1] in ["Josa", 'Eomi', 'Punctuation']:
            josa_list.append(word[0])
            
print(noun_list)
print("==================================================================================================")
print(verb_list)
print("==================================================================================================")
print(josa_list)
```

```python
noun_list_su=[]
verb_list_su=[]
josa_list_su=[]

for w in df_수정결과["Script"]:
    for word in okt.pos(w):
        if word[1] in ["Noun"]:
            noun_list_su.append(word[0])
        if word[1] in ["Verb"]:
            verb_list_su.append(word[0])
        if word[1] in ["Josa", 'Eomi', 'Punctuation']:
            josa_list_su.append(word[0])
            
print(noun_list_su)
print("==================================================================================================")
print(verb_list_su)
print("==================================================================================================")
print(josa_list_su)
```

```python
noun_num=0
for i in noun_list:
    if i not in noun_list_su:
        noun_num+=1
        
if noun_num==0:
    print("일치하므로 accuracy는 100%입니다")
else:
    print(noun_num)
```
